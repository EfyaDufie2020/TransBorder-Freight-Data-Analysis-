### Features

   #### **1. Shipment Identification & Trade Information**  
| **Feature**     | **Description**  |  
|----------------|--------------------------------------------------------------------------------------------------|  
| **YEAR**       | **Year** of the shipment (four-digit AD format). |  
| **MONTH**      | **Month** of the shipment (1 - 12). |  
| **TRDTYPE**    | Trade type, indicating whether the shipment is an **(1 =export)** or **(2=import)**. |  
| **COMMODITY2** | **2-digit commodity code** categorizing the type of goods being transported. |  
 
#### **2. Shipment Origin & Destination Details**  
| **Feature**   | **Description**  |  
|--------------|--------------------------------------------------------------------------------------------------|  
| **USASTATE**  | U.S. state code where the freight originates or arrives. |  
| **MEXSTATE**  | Mexican state code, applicable when the shipment involves Mexico. |  
| **CANPROV**   | Canadian province code, applicable when the shipment involves Canada. |  
| **COUNTRY**   | Country code indicating the international origin or destination of the shipment (Canada: **1220**, Mexico: **2010**). |  
| **DEPE**      | Port or district code representing the shipment's processing location. |  
 
#### **3. Transportation & Logistics Information**  
| **Feature**    | **Description**  |  
|--------------|--------------------------------------------------------------------------------------------------|  
| **DISAGMOT**  | Mode of transportation code specifying how the freight is transported (**1 = Vessel, 3 = Air, 4 = Mail, 5 = Truck, 6 = Rail, 7 = Pipeline, 8 = Other, 9 = Foreign Trade Zones (FTZs)**). |  
| **CONTCODE**  | Indicates whether the shipment is **containerized (X)** or **non-containerized (0)**. |  
 
#### **4. Economic & Cost Factors**  
| **Feature**         | **Description**  |  
|--------------------|--------------------------------------------------------------------------------------------------|  
| **VALUE**          | Total **value of goods** being shipped, measured in U.S. dollars (USD). |  
| **SHIPWT**         | Total **shipping weight** of the goods in kilograms (Kg). |  
| **FREIGHT_CHARGES** | **Freight cost** associated with transporting the shipment (in USD). |  
| **DF**             | Indicates whether the **merchandise was produced in the U.S. (1 = Domestic) or outside the U.S. (2 = Foreign)**. |



   
### Hypothesis
The trade patterns, transportation modes, and commodity types significantly influence the value and volume of cross-border freight between the United States, Canada, and Mexico.

- Null Hypothesis (H0): There is no significant relationship between transportation modes and value of cross-border freight.

- Alternative Hypothesis (H1): There is a significant relationship between transportation modes and the value of cross-border freight.



### Analytical Questions
Here are 7 analytical questions for the **transborder freight analysis** based on my hypothesis: 

1. Which transportation mode is most commonly used for cross-border freight?

2. How has the total freight value changed over time?

3. What are the top 10 U.S. states importing and exporting goods?

4. Which country (Canada/Mexico) receives more freight?

5. What is the correlation between freight weight and value?

6. What are the top 5 commodities (arrange from highest to lowest) with the highest total shipment value across the United States, Canada, and Mexico, and how do they compare to each other in terms of value?

7. What is the distriution of the total shipment values for the Containerized and Non-Containerized Shipment?

8. What is the distriution of the total shipment values for the domestically produced and foreign produced merchandise?

9. What is the distriution of the total shipment values for the domestically produced and foreign produced merchandise?

10. What is the total value of the shipment distribution across the USASTATE, MEXSTATE & CANPROV?

### Hypothesis Testing

```python
# Check the distribution of the 'VALUE' column
value_distribution = new_df['VALUE'].hist()
plt.show()

# Perform the appropriate statistical test
transportation_modes = new_df['DISAGMOT'].unique()

# Check if the 'VALUE' column is normally distributed
alpha = 0.05
k2, p = stats.normaltest(new_df['VALUE'])

if p < alpha:
    # 'VALUE' column is not normally distributed, use Kruskal-Wallis H test
    statistic, p_value = stats.kruskal(*[new_df[new_df['DISAGMOT'] == mode]['VALUE'] for mode in transportation_modes])
else:
    # 'VALUE' column is normally distributed, use ANOVA test
    f_value, p_value = stats.f_oneway(*[new_df[new_df['DISAGMOT'] == mode]['VALUE'] for mode in transportation_modes])

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis (H0). There is a significant relationship between transportation modes and the value of cross-border freight.")
else:
    print("Fail to reject the null hypothesis (H0). There is no significant relationship between transportation modes and the value of cross-border freight.")
```

#### Key Observations & Comments on the Hypothesis Testing & Univariate Analysis

1. Distribution of VALUE (Freight Value)

-Observation:

The histogram of VALUE suggests that the distribution is right-skewed (positively skewed), meaning most shipments have lower values, while a few high-value shipments create a long right tail.
This skewness is common in freight data, where bulk commodity shipments (e.g., raw materials) dominate in volume but not in value, while high-value shipments (e.g., electronics, pharmaceuticals) are less frequent but more expensive.

- Comment:

The presence of extreme values (outliers) could impact statistical analysis, especially tests assuming normality.
Further log-transformation or normalization may help improve model performance when using value-based predictive analytics.

2️. Normality Test for VALUE

- Observation:

The D’Agostino-Pearson normality test (stats.normaltest) checks whether VALUE follows a normal distribution.
The p-value is below 0.05, confirming that the VALUE column is not normally distributed.

-Comment:

Since VALUE is non-normally distributed, parametric tests like ANOVA (which assumes normality) cannot be used.
Instead, a non-parametric Kruskal-Wallis H test is used to compare VALUE across transportation modes.

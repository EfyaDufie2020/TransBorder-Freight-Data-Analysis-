## TRANSBORDER FRIEGHT DATA ANALYSIS

## Business Understanding

### Background
The project aims to leverage data analysis to extract actionable insights that address key business objectives. By implementing the CRISP-DM (Cross-Industry Standard Process for Data Mining) framework, this analysis will systematically explore the dataset provided and deliver value through insightful visualizations and actionable recommendations.

### Objective
The primary objective of this project is to analyze the given dataset to uncover trends, patterns, and insights that can guide informed decision-making. These insights will serve as the foundation for actionable business strategies that align with the overall goals of the organization.

### Goals

Develop a deep understanding of the dataset to ensure robust and accurate analysis.
Identify key analytical questions that will drive the exploration and insights.
Use appropriate tools and techniques to preprocess the data, extract insights, and visualize the findings effectively.
Document the entire process for reproducibility, transparency, and future reference.

### Scope
The project will include:

### Hypothesis
The trade patterns, transportation modes, and commodity types significantly influence the value and volume of cross-border freight between the United States, Canada, and Mexico.

- Null Hypothesis (H0): There is no significant relationship between trade patterns, transportation modes, commodity types, and the value/volume of cross-border freight.

- Alternative Hypothesis (H1): There is a significant relationship between trade patterns, transportation modes, commodity types, and the value/volume of cross-border freight.



### Analytical Questions
1. How do the trade values and shipping weights vary across different modes of transportation (DISAGMOT) for imports and exports (TRDTYPE) between the United States, Canada, and Mexico?

2. What are the top 5 commodities (COMMODITY2) traded in terms of value and weight, and how do they differ between Canada and Mexico?
3. Is there a significant difference in freight charges (FREIGHT_CHARGES) for containerized (CONTCODE) versus non-containerized shipments across different transportation modes?
4. How have the trade patterns (in terms of value and volume) changed over the years (2020-2024) for domestic (DF=1) versus foreign (DF=2) produced merchandise?

5. Which U.S. states (USASTATE) have the highest trade values with specific Mexican states (MEXSTATE) and Canadian provinces (CANPROV), and what are the primary commodities involved in these trades?
